### 제대로 안긁어와졌음. 통째로 긁어와서 넣어야 할듯. 

```python
# 크롬 브라우저를 열기
DRIVER_DIR = '../driver/chromedriver'

driver = webdriver.Chrome(DRIVER_DIR)
```

```python
# 목표 페이지 불러오기
index = "https://outschool.com/online-classes/" 

theme_list = ["arts","coding-tech"]
#theme_list = ["arts","coding-tech","english","health-wellness","life-skills","math","music","science-nature", "social-studies", "world-languages"]

df = []
data = []

for i in theme_list:
    url = index + i
    driver.get(url)
    
    time.sleep(6)
    
    
    # div 소스 가져오기
    theme = driver.find_elements_by_tag_name('h1')[0].text

    html = driver.page_source
    bsObject = BeautifulSoup(html, 'lxml')
    div_classes = bsObject.find_all('div', {'class':'card row'})
        
        
    # 상세 항목 추출하기
    for div in div_classes:
        try:
            title = div.find('div', {'class':'css-xihznx-Flex'}).get_text()  
            review_num = div.find('div', {'class':'card-ribbon'}).get_text()        
            desc = div.find('div', {'class':'half-space-above flex-centered-row flex-wrap darkgray'}).get_text()
        except Exception:
            continue
            
        data.append((theme, title, review_num, desc))
        
    
print('well done')
```

```python
# 데이터프레임으로 변환하기

df = pd.DataFrame(data)
df.columns = ('theme','title','review_num', 'desc')

df.shape
```

### 영상 제목 처리

```python
# 영상 회차 컬럼 생성하기
df['title_no'] = df['title'].str.split(" ").str[3].str.replace("강", "")
df['title_no'] = pd.to_numeric(df['title_no'])
df.head()
```

```python
# 챕터 매칭하기
def match_chapter(title_no):
    if title_no <= 6:
        return "1_시작하기"
    elif 7 <= title_no <= 12:
        return "2_자료형"
    elif 13 <= title_no <= 17:
        return"3_조건문"
    elif 18 <= title_no <= 30:
        return "4_반복문"
    elif 31 <= title_no <= 40:
        return "5_함수"
    elif 41 <= title_no <= 44:
        return "6_예외처리_중급"
    elif 45 <= title_no <= 48:
        return "7_클래스_고급"
    else:
        return "8_모듈_고급"
    
df['chapter'] = df['title_no'].apply(match_chapter)
```

```python
# 제목만 추출하여 컬럼 생성하기
df['title_desc'] = df['title'].str.split(" - ").str[1]
df
```

### 시간 전처리

```python
# 시간에서 시/분을 추출하여 컬럼 생성
df['time_min'] = df['duration'].str.split(":").str[0]
df['time_sec'] = df['duration'].str.split(":").str[1]

df.head()
```

```python
# 시/분을 숫자형으로 변환
df['time_min'] = pd.to_numeric(df['time_min'])
df['time_sec'] = pd.to_numeric(df['time_sec'])

df.info()
```

```python
# 재생시간을 분으로 표현 (초는 소수점 처리)
df['time_min_total'] = (df['time_min'] + df['time_sec']/60)

# 소수점 2자리까지만 표현하도록 셋팅 
# (참고자료 : https://financedata.github.io/posts/pandas-display-format.html)
pd.options.display.float_format = '{:.2f}'.format
df.head()
```

```python
# 파일로 저장하기
df.to_csv("self_study_python_youtube.csv", index=False)
```

## 3. 파악 및 분석

```python
# 전체를 공부하는 데 걸리는 재생 시간 : 약 10시간
minute = df['time_min_total'].sum() // 1
second = ((df['time_min_total'].sum() % 1) * 60).round(0)

print("전체 재생시간 : {}분 {}초".format(int(minute), int(second)))
```

```python
# duration을 시간 포맷으로 변경
df['duration'] = pd.to_datetime(df['duration'], format='%M:%S')
df.info()
```
