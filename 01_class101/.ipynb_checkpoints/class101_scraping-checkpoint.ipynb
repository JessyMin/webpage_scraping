### Lesson & Learn
 - 사용한 방법 : 속성값 별로 긁어와서 리스트를 생성한 뒤 데이터프레임으로 합치는 방식
 - 문제점 : 데이터에 Null값이 있을 경우, 리스트의 길이가 맞지 않게 됨
 - 리뷰 단위로 긁어와서 속성을 뽑아내는 방식으로 해야 함

```python
# 크롬 브라우저를 열기
from selenium import webdriver

DRIVER_DIR = 'driver/chromedriver'

chrome = webdriver.Chrome(DRIVER_DIR)
```

```python
# 목표 페이지 불러오기
url = "https://class101.net/categories/5e954c8389bffe06ef955686"

chrome.get(url)
```

```python
# 크리에이터 크롤링 

creator = []

creators = chrome.find_elements_by_class_name('kPPRIZ')

for i in creators:
    creator.append(i.text)
    
print(creator)
len(creator)
```

```python
# 강의 제목 크롤링 

title = []

titles = chrome.find_elements_by_class_name('sc-fzoJMP')

for i in titles:
    title.append(i.text)

print(title)
len(title)
    
```

```python
# 좋아요 수 크롤링 
favorite = []

xpath='/html/body/div[1]/div/div[1]/main/div/div/div[2]/div/div/div/ul/li/a/div/div[2]/div[3]/div/div/div[1]'
favorites = chrome.find_elements_by_xpath(xpath)

for i in favorites:
    favorite.append(i.text)

print(favorite)
len(favorite)
    
```

```python
# 만족도 크롤링 
like = []

xpath='/html/body/div[1]/div/div[1]/main/div/div/div[2]/div/div/div/ul/li/a/div/div[2]/div[3]/div/div/div[2]'
likes = chrome.find_elements_by_xpath(xpath)

for i in likes:
    like.append(i.text)

print(like)
len(like)
    
```

```python
# 데이터프레임으로 저장하기
import pandas as pd

zippedList = list(zip(creator, title, favorite, like))
df= pd.DataFrame(zippedList, columns = ['creator', 'title', 'favorite', 'like'])

df.head(10)
```

```python
# 드라이버 종료
chrome.quit()
```
