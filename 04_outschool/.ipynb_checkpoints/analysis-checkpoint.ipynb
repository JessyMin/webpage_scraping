```python
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
# 숫자로 변환하기
df['review_num'].str.replace("New Class","0")

df['review_num'] = pd.to_numeric(df['review_num'], errors="coerce")
```
